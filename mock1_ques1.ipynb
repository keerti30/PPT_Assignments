```python
def movezeros(nums):
    i=0
    j=0

    while i < len(nums):
        if nums[i]!=0:
            nums[i],nums[j]=nums[j],nums[i]
            j+=1
        i+=1
    while j< len(nums):
        nums[j]=0
        j+=1
    return nums
```

```python
nums = [0,1,0,3,12]
result = movezeros(nums)
print(result)
```

```python
nums = [0]
result = movezeros(nums)
print(result)
```
